# **ANÁLISE FINANCEIRA DO CLUBE DE FUTEBOL E REGATAS DO FLAMENGO**

Importando dados

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

Definindo Dados e criando DF

```python

# Definir os dados
data = {
    'Ano': [2023, 2022, 2021, 2020],
    'TOTAL DO ATIVO CIRCULANTE': [455917, 343558, 304754, 198226],
    'TOTAL DO ATIVO NÃO CIRCULANTE': [874035, 788335, 761873, 746296],
    'TOTAL DO PASSIVO CIRCULANTE': [375282, 392167, 366071, 495065],
    'TOTAL DO PASSIVO NÃO CIRCULANTE': [360529, 405180, 501700, 428248],
    'ESTOQUE': [3780, 4113, 3391, 2299],
    'TOTAL DO PATRIMÔNIO LÍQUIDO': [654091, 334546, 198856, 21209],
    'RECEITA OPERACIONAL LÍQUIDA': [1315873, 1112429, 1025428, 644093],
    'RESULTADO BRUTO': [379789, 284891, 263818, 29145],
    'RESULTADO OPERACIONAL': [291698, 156676, 188145, -65614]
}

# Criar o DataFrame
df = pd.DataFrame(data)


```

Calculando indicadores

```python
# Calcular os Indicadores Financeiros
df['LIQUIDEZ CORRENTE'] = df['TOTAL DO ATIVO CIRCULANTE'] / df['TOTAL DO PASSIVO CIRCULANTE']
df['LIQUIDEZ SECA'] = (df['TOTAL DO ATIVO CIRCULANTE'] - df['ESTOQUE']) / df['TOTAL DO PASSIVO CIRCULANTE']
df['MARGEM DE LUCRO BRUTA'] = df['RESULTADO BRUTO'] / df['RECEITA OPERACIONAL LÍQUIDA'] * 100
df['MARGEM DE LUCRO OPERACIONAL'] = df['RESULTADO OPERACIONAL'] / df['RECEITA OPERACIONAL LÍQUIDA'] * 100
df['ÍNDICE DE ENDIVIDAMENTO'] = (df['TOTAL DO PASSIVO CIRCULANTE'] + df['TOTAL DO PASSIVO NÃO CIRCULANTE']) / df['TOTAL DO PATRIMÔNIO LÍQUIDO']

# Exibir o DataFrame com os Indicadores Financeiros
print(df)

```

Criando Gráficos para Apresentação de Resultados

```python
# Gerar Gráficos
plt.figure(figsize=(14, 10))

# Gráfico de Liquidez Corrente e Liquidez Seca
plt.subplot(2, 3, 1)
plt.plot(df['Ano'], df['LIQUIDEZ CORRENTE'], marker='o', label='Liquidez Corrente')
plt.plot(df['Ano'], df['LIQUIDEZ SECA'], marker='o', label='Liquidez Seca')
plt.title('Liquidez Corrente e Liquidez Seca')
plt.xlabel('Ano')
plt.ylabel('Razão')
plt.xticks(df['Ano'])  # Corrigir rótulos do eixo x
plt.legend()

# Gráfico de Margem de Lucro Bruta
plt.subplot(2, 3, 2)
plt.plot(df['Ano'], df['MARGEM DE LUCRO BRUTA'], marker='o', color='g')
plt.title('Margem de Lucro Bruta (%)')
plt.xlabel('Ano')
plt.ylabel('Percentual')
plt.xticks(df['Ano'])  # Corrigir rótulos do eixo x



# Gráfico de Margem de Lucro Operacional
plt.subplot(2, 3, 3)
plt.plot(df['Ano'], df['MARGEM DE LUCRO OPERACIONAL'], marker='o', color='r')
plt.title('Margem de Lucro Operacional (%)')
plt.xlabel('Ano')
plt.ylabel('Percentual')
plt.xticks(df['Ano'])  # Corrigir rótulos do eixo x

# Gráfico de Índice de Endividamento
plt.subplot(2, 3, 4)
plt.plot(df['Ano'], df['ÍNDICE DE ENDIVIDAMENTO'], marker='o', color='m')
plt.title('Índice de Endividamento')
plt.xlabel('Ano')
plt.ylabel('Razão')
plt.xticks(df['Ano'])  # Corrigir rótulos do eixo x

plt.tight_layout()
plt.show()
```
